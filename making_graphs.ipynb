```python
import matplotlib.pyplot as plt
```

```python
loss = [0.2656,0.2648,0.2621,.2609,0.2591,0.2582,0.2552,0.2548,0.2529,0.2522]
len(loss)
```

```python
train_acc = [96.14,96.13,96.28,96.40,96.48,96.52,96.73,96.78,96.87,96.92]
len(train_acc)
```

```python
val_acc = [94.89,95.52,93.81,94.34,96.15,95.76,95.97,95.82,94.22,94.92]
len(val_acc)
```

```python
epochs = range(1, len(train_acc) + 1)

plt.figure(figsize=(6, 4), dpi=300)
plt.plot(epochs, train_acc, label="Training Accuracy")
plt.plot(epochs, val_acc, label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.tight_layout()
plt.savefig("accuracy_plot.png", dpi=300)
```

```python
plt.figure(figsize=(6, 4), dpi=300)
plt.plot(epochs, loss, label="Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig("loss_plot.pdf")
plt.savefig("loss_plot.png", dpi=300)
```
